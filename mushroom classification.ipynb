```python
df.drop(['veil-type'], axis= 1,inplace= True)
```

```python
df['class'].value_counts()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('class', axis= 1)
y = df['class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
transformer = ColumnTransformer(transformers=[
    ('tnf1', OneHotEncoder(sparse= False, drop='first'),df.columns.drop('class'))
], remainder= 'passthrough')
```

```python
transformed_xtrain = transformer.fit_transform(X_train)
transformed_xtest = transformer.transform(X_test)
```

```python
lencoder = LabelEncoder()
transformed_ytrain = lencoder.fit_transform(y_train)
transformed_yest = lencoder.transform(y_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
def run_model(model):
  model.fit(transformed_xtrain, transformed_ytrain)
  preds = model.predict(transformed_xtest)
  print("Confusion Matrix:  ")
  print(confusion_matrix(transformed_yest, preds))
  print("---"*10)
  print("Accuracy Score is: ", accuracy_score(transformed_yest,preds))
  print("---"*10)
  print("Confusion Matrix:  ")
  print(classification_report(transformed_yest, preds))
```

```python
model1  = LogisticRegression()

run_model(model1)
```
